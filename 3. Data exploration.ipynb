```python
"""
Subcass PrettyPrinter to print UTF-8 characters
"""
class MyPrettyPrinter(pprint.PrettyPrinter):
    def format(self, object, context, maxlevels, level):
        if isinstance(object, unicode):
            return (object.encode('utf8'), True, False)
        return pprint.PrettyPrinter.format(self, object, context, maxlevels, level)
```

```python
pp = pprint.PrettyPrinter()
mypp = MyPrettyPrinter()
```

```python
# MongoDB constants
DATABASE_LOC = "mongodb://127.0.0.1:27017"
DATABASE_NAME = "hk_osm"
COLLECTION_HEADS = "heads"
COLLECTION_ELEMS = "elems"
COLLECTION_USERS = "users"
KEY_CREATED = "created"
KEY_LOC = "loc"
# OSM-specific constants
HONGKONG_OSM = "hong-kong_china.osm"
TAG_BOUNDS = "bounds"
TAG_NODE = "node"
TAG_WAY = "way"
TAG_RELATION = "relation"
ELEMENTS_HEADER = [TAG_BOUNDS]
ELEMENTS_CORE = [TAG_NODE, TAG_WAY, TAG_RELATION]
ATTR_VERSION = "version"
ARRT_CHANGESET = "changeset"
ATTR_TIMESTAMP = "timestamp"
ATTR_USER = "user"
ATTR_UID = "uid"
ATTR_LON = "lon"
ATTR_LAT = "lat"
ATTRIBUTES_CREATED = [ATTR_VERSION, ARRT_CHANGESET, ATTR_TIMESTAMP, ATTR_USER, ATTR_UID]
ATTRIBUTES_LOC = [ATTR_LON, ATTR_LAT]
```

```python
"""
Initialize connection, clear up user collections and return the database
"""
def init_database(loc, name):
    from pymongo import MongoClient
    client = MongoClient(loc)
    
    db = client[name]
    pp.pprint(client[name])

    return db
```

```python
db = init_database(DATABASE_LOC, DATABASE_NAME)
```

```python
"""
Find the top contributors and the total count of elements created by them
"""
def get_count_by_top_contributors(limit, display=False):
    pipeline = [{"$group": {"_id": "$_id", "totalCount": {"$sum": "$count"}}},
                {"$sort": {"totalCount": -1}},
                {"$limit": limit}]
    contributors = [i for i in db[COLLECTION_USERS].aggregate(pipeline)]
    
    if display:
        print "Top {0} contributors:".format(limit)
        mypp.pprint(contributors)
        
    topTotalCount = 0
    for i in contributors:
        topTotalCount += int(i["totalCount"])
    
    return topTotalCount
```

```python
def check_top_contributors():
    userCount = db[COLLECTION_USERS].count()
    print "Total count of unique contributors: {0}".format(userCount)
    elementCount = db[COLLECTION_ELEMS].count()
    print "Average element count per contributor: {0:.2f}".format(float(elementCount)/userCount)
    
    percentage = 0.01
    limit = int(userCount * 0.01 + 0.5)
    print "Contribution of top {0:.0%} contributors: {1:.2%}".format(percentage, float(get_count_by_top_contributors(limit, display=True))/elementCount)
    percentage = 0.1
    limit = int(userCount * 0.1 + 0.5)
    print "Contribution of top {0:.0%} contributors: {1:.2%}".format(percentage, float(get_count_by_top_contributors(limit))/elementCount)
    
check_top_contributors()
```

```python
"""
Check the top values filled in for the specified feature key
"""
def check_top_feature_values(elemTypes, key, limit):
    pipeline = [{"$match": {"type": {"$in": elemTypes}, "tags.k": key}},
                {"$project": {"tags": 1}},
                {"$unwind": "$tags"},
                {"$match": {"tags.k": key}},
                {"$group": {"_id": "mock", "count": {"$sum": 1}}}]
    featureCount = db[COLLECTION_ELEMS].aggregate(pipeline).next()["count"]
    print "{0} feature tags of {1} were logged".format(featureCount, key)
    
    pipeline = [{"$match": {"type": {"$in": elemTypes}, "tags.k": key}},
                {"$project": {"tags": 1}},
                {"$unwind": "$tags"},
                {"$match": {"tags.k": key}},
                {"$group": {"_id": "$tags.v", "count": {"$sum": 1}}},
                {"$sort": {"count": -1}},
                {"$limit": limit}]
    values = [i for i in db[COLLECTION_ELEMS].aggregate(pipeline)]
    print "Top {0} values of {1}:".format(limit, key)
    mypp.pprint(values)
```

```python
import re
# regular expressions
REGX_DOOR = re.compile(r"^[0-9]+[A-E][0-9]+$")
REGX_DOOR_MULTI = re.compile(r"^[0-9]+[A-E]([0-9\s,-])*$")
REGX_BLOCK = re.compile(r"[A-E]")
REGX_FLOOR = re.compile(r"^([0-9]+/F)+(-[0-9]+/F|)$")
REGX_HOUSENUM = re.compile(r"^Block [A-E]")
```

```python
"""
Helper function to add/update hotel stats
"""
def add_hotel(hotel_stats, block, floor, name):
    if block not in hotel_stats:
        hotel_stats[block] = {"count": 0}
    if floor not in hotel_stats[block]:
        hotel_stats[block][floor] = {"count": 0, "hotels": []}
    if name:
        hotel_stats[block]["count"] += 1
        hotel_stats[block][floor]["count"] += 1
        hotel_stats[block][floor]["hotels"].append(name)
```

```python
"""
A famous residential building:

Chungking Mansions is well known to tourists as nearly the cheapest accommodation in Hong Kong. Though the building is 
supposedly residential, it is made up of many independent low-budget hotels, shops and other services. The unusual atmosphere 
of the building is sometimes compared to that of the former Kowloon Walled City.
http://en.wikipedia.org/wiki/Chungking_Mansions

On the map it seems someone was logging all the cheap hotels (guesthouses) in the complex. I'm exploring those programmatically.
"""

def explore_chungking():
    building = db[COLLECTION_ELEMS].find_one({"tags.k": "name", 
                                              "tags.v": {"$in": ["Chungking Mansion", "Chungking Mansions", "重慶大廈"]}})
    mypp.pprint(building)
    # it's a way element, getting the nds the outline this complex
    pipeline = [{"$match": {"type": "node", "id": {"$in": building["nds"]}}},
                {"$group": {"_id": "nd", "locs": {"$push": "$loc"}}}]
    polygon = [i for i in db[COLLECTION_ELEMS].aggregate(pipeline)][0]["locs"]
    # find possible touristic accommodations in the complex
    cursor = db[COLLECTION_ELEMS].find({"type": "node", 
                                        "tags.k": "tourism", 
                                        "tags.v": {"$in": ["apartment", "guest_house", "hostel", "hotel", "motel"]}, 
                                        "loc": {"$geoWithin": {"$polygon": polygon}}})
    # convert to list for operation convenience
    hotels = [i for i in cursor]   
    hotel_stats = {}
    problem_doors = []
    problem_hotels = []
    # I noticed that the contibutor accustomed to use this format to log business in the apartment unit {k: addr:door, v: 9E5}
    # 9E5 stands for: unit #5, 9th floor, block E (Chungking Mansions is consisted of 5 blocks A, B, C, D and E)
    # Such element constantly comes with these tags: {k: level, v: 9}
    #                                                {k: addr:floor, v: 9/F}
    #                                                {k: addr:street, v: 彌敦道 Nathan Road}
    #                                                {k: addr:housenumber, v: Block E, 36-44}
    # These can be used for cross checking, correcting and data enhancement
    # Below I'm focusing on addr:door tags to see how contributers like to log the information
    # And it looks like some guesthouses span across multiple units horizontally or vertically
    for h in hotels:
        hasAddrDoor = False
        for t in h["tags"]:
            if t["k"] == "name":
                name = t["v"]
            elif t["k"] == "addr:door":
                hasAddrDoor = True
                v = t["v"]
                m = REGX_DOOR.search(v)
                if m:
                    m = REGX_BLOCK.search(v)
                    block = m.group(0)
                    preAndSuff = v.split(block)
                    if preAndSuff and len(preAndSuff) > 0:
                        floor = int(v.split(block)[0])
                        add_hotel(hotel_stats, block, floor, name)
                else: # no match
                    problem_doors.append(v)
                    problem_hotels.append(h)
                    break
        if hasAddrDoor == False:
            problem_hotels.append(h)

    print "Chungking Mansion has a total of {0} hotels".format(len(hotels))
    print "{0} hotels with usual door numbers".format(len(hotels)-len(problem_hotels))
    print "{0} hotels with missing/unusual door numbers".format(len(problem_hotels))
    print "Hotels with usual door numbers"
    mypp.pprint(hotel_stats)
    
    return hotels
```

```python
"""
I noticed that the contributor accustomed to use this format to log business in the apartment unit {k: addr:door, v: 9E5}
9E5 stands for: unit #5, 9th floor, block E (Chungking Mansions is consisted of 5 blocks A, B, C, D and E)
Such element constantly comes with these tags: {k: level, v: 9}
                                               {k: addr:floor, v: 9/F}
                                               {k: addr:street, v: 彌敦道 Nathan Road}
                                               {k: addr:housenumber, v: Block E, 36-44}
Below I'm doing a quick completeness and cross checking
And it looks like some guesthouses span across multiple units horizontally or vertically

For furthur exploration and development, I can scrap data from their official website http://www.chungking-mansions.hk/index.html and
import more data into OpenStreetMap. But for now, I'll just stay with what's available in OSM. 
"""
def check_addr_chungking(hotels):
    addr_stats = {"addr:housenumber": {},
                  "level_floor": {},
                  "addr:door_special": []}
    # extract values
    for h in hotels:
        housenumber = None
        door = None
        level = None
        floor = None
        for t in h["tags"]:
            v = t["v"]
            k = t["k"]
            if k == "addr:housenumber":
                housenumber = v
                if v not in addr_stats["addr:housenumber"]:
                    addr_stats["addr:housenumber"][v] = 0
                addr_stats["addr:housenumber"][v] += 1
            elif k == "level":
                level = v
            elif k == "addr:floor":
                m = REGX_FLOOR.search(v)
                if m:
                    floor = v.replace("-", "").split("/F")
                    floor = ",".join(i for i in floor[:len(floor)-1])
                else:
                    floor = v
            elif k == "addr:door":
                door = v
                m = REGX_DOOR.search(v)
                if not m and v not in addr_stats["addr:door_special"]:
                    addr_stats["addr:door_special"].append(v)
         # cross checking 
        if not door:
            key = "missing(housenumber={0}, level={1}, floor={2})".format(housenumber, level, floor)
            if key not in addr_stats["addr:door_special"]:
                addr_stats["addr:door_special"].append(key)     
        if not housenumber:
            key = "missing(door={})".format(door)
            if key not in addr_stats["addr:housenumber"]:
                addr_stats["addr:housenumber"][key] = 0
            addr_stats["addr:housenumber"][key] += 1
        if not level:
            level = "missing(door={})".format(door)
        if not floor:
            floor = "missing(door={})".format(door)
        l_f = "{0}_{1}".format(level, floor)
        if l_f not in addr_stats["level_floor"]:
            addr_stats["level_floor"][l_f] = 0
        addr_stats["level_floor"][l_f] += 1
                
    mypp.pprint(addr_stats)
```

```python
chungking_hotels = explore_chungking()
print "\n\n"
check_addr_chungking(chungking_hotels)
```

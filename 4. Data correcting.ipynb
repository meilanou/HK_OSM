```python
import pprint
import pymongo
import re
# MongoDB constants
DATABASE_LOC = "mongodb://127.0.0.1:27017"
DATABASE_NAME = "hk_osm"
COLLECTION_HEADS = "heads"
COLLECTION_ELEMS = "elems"
COLLECTION_USERS = "users"
KEY_CREATED = "created"
KEY_LOC = "loc"
# regular expressions
REGX_DOOR = re.compile(r"^[0-9]+[A-E][0-9]+$")
REGX_DOOR_MULTI = re.compile(r"^[0-9]+[A-E]([0-9\s,-])*$")
REGX_BLOCK = re.compile(r"[A-E]")
REGX_FLOOR = re.compile(r"^([0-9]+/F)+(-[0-9]+/F|)$")
REGX_HOUSENUM = re.compile(r"^Block [A-E]")
```

```python
"""
Subcass PrettyPrinter to print UTF-8 characters
"""
class MyPrettyPrinter(pprint.PrettyPrinter):
    def format(self, object, context, maxlevels, level):
        if isinstance(object, unicode):
            return (object.encode('utf8'), True, False)
        return pprint.PrettyPrinter.format(self, object, context, maxlevels, level)
```

```python
pp = pprint.PrettyPrinter()
mypp = MyPrettyPrinter()
```

```python
"""
Initialize connection, clear up user collections and return the database
"""
def init_database(loc, name):
    from pymongo import MongoClient
    client = MongoClient(loc)
    
    db = client[name]
    pp.pprint(client[name])

    return db
```

```python
db = init_database(DATABASE_LOC, DATABASE_NAME)
```

```python
"""
The goal is to extract values from tag elements and add as new dictionary items for quick access. The new format will be:

name: Comfort Guesthouse,
addr: {block: E,
        floor: 9,
        housenumber: Block E, 36-44,
        room: 5,
        street: 彌敦道 Nathan Road}
        
In Hong Kong "room" is used to mark apartment units in a residential building. Although for OSM, "room" is used to 
mark the purpose of a space in the building. The contributor used "addr:door" for apartment units and I'll extract it to 
addr-room and make it more harmonized (9E5 -> 5, 15B1,6,8 -> 1,6,8). And contributors like to put block info with addr:housenumber. 
Since "block" is commonly used in Hong Kong, I'll extract block and put it in addr-block (Block A -> A).

Contributors use suffix "/F" for addr:floor tags and it's redundant. I'll remove "/F" and just use numbers.

And here contributors are mix-using level and addr:floor. Level is supposed to be floor-1. I'll keep the values for local custom. 
If a guesthouse spans vertically, since I don't know how it's configured I'll use multiple values for addr-floor (4/F-5/F -> 4,5).

addr:unit - The number, letter, or name of a single unit or flat that exists within a larger complex. While a big building 
could have only one entrance, sometimes the way inside divides into different units or staircases, where certain 
apartments/flats/offices can only be reached though a specific unit. Useful for apartment, suite, or office numbers. 
Information necessary for postmen, for example.

addr:door - The door number (or name) of an apartment/flat/office/room. Could also be called differently in some countries 
(In Austria for example it is sometimes referred to as "Top"). If a room has more than one door, you can tag them separately.

addr:floor - The floor an apartment/flat/office is located. 

level - Multiple values level=1 floor over ground floor (British English: first floor, American English: second floor). 
Level range: level=-1;0 room=yes is a room which goes over two levels. Often there are doors to each one.
"""
def correct_chungking():
    building = db[COLLECTION_ELEMS].find_one({"tags.k": "name", "tags.v": {"$in": ["Chungking Mansion", "Chungking Mansions", "重慶大廈"]}})
    #mypp.pprint(building)
    # it's a way element, getting the nds the outline this complex
    pipeline = [{"$match": {"type": "node", "id": {"$in": building["nds"]}}},
                {"$group": {"_id": "nd", "locs": {"$push": "$loc"}}}]
    polygon = [i for i in db[COLLECTION_ELEMS].aggregate(pipeline)][0]["locs"]
    # find possible touristic accommodations in the complex
    cursor = db[COLLECTION_ELEMS].find({"type": "node", 
                                        "tags.k": "tourism", 
                                        "tags.v": {"$in": ["apartment", "guest_house", "hostel", "hotel", "motel"]}, 
                                        "loc": {"$geoWithin": {"$polygon": polygon}}})
    # convert to list for operation convenience
    hotels = [i for i in cursor] 
    update_docs = []

    for h in hotels:
        node_id = None
        name = None
        street = None
        housenumber = None
        block_house = None
        block_door = None
        door = None
        level = None
        floor = None
        floor_door = None
        node_id = h["id"]
        for t in h["tags"]:
            k = t["k"]
            v = t["v"]
            if k == "name":
                name = v
            elif k == "addr:street":
                street = v
            elif k == "addr:housenumber":
                housenumber = v
                m = REGX_HOUSENUM.search(housenumber)
                if m:
                    block_house = REGX_BLOCK.search(m.group(0)).group(0)
            elif k == "level":
                level = v
            elif k == "addr:floor":
                m = REGX_FLOOR.search(v)
                if m:
                    floor = v.replace("-", "").split("/F")
                    floor = ",".join(i for i in floor[:len(floor)-1])
            elif k == "addr:door":
                isMulti = False
                door = v.replace(" ", "")
                m = REGX_DOOR.search(door)
                if not m:
                    m = REGX_DOOR_MULTI.search(door)
                    if m:
                        isMulti = True
                if m:
                    block_door = REGX_BLOCK.search(m.group(0)).group(0)
                    items = door.split(block_door)
                    floor_door = items[0]
                    door = items[1]
                    if isMulti:
                        items = door.split("-")
                        if len(items) > 1: # len = 2
                            door = []
                            for i in range(int(items[0]), int(items[1])+1):
                                door.append(i)
                            door = ",".join(str(i) for i in door)
                            
        # according to data exploration, examples to be corrected
        # 2938367944 Block A, 36-44 B    None None  4 4,5  None
        # 2937037495 None           None D    8     7 None 7
        # 2936536786 Block B, 36-44 B    B    1,3,7 7 7    7
        block = block_door or block_house
        floor = floor or floor_door
        level = floor
        if not block_house:
            block_house = block_door
        if not housenumber:
            housenumber = "Block {}, 36-44".format(block)
        doc = {"filter": {"id": node_id, "type": "node"},
               "update": {"$set": {"name": name, 
                                   "addr": {"street": street,
                                            "housenumber": housenumber,
                                            "block": block,
                                            "floor": floor}}}}
        if door:
            doc["update"]["$set"]["addr"]["room"] = door
        update_docs.append(doc)
        
    matched_count = 0
    modified_count = 0
    for d in update_docs:
        result = db[COLLECTION_ELEMS].update_one(d["filter"], d["update"], upsert=True)
        matched_count += result.matched_count
        modified_count += result.modified_count
    print "{0} documents prepared for update. {1} found for update, and {2} updated.".format(len(update_docs), matched_count, modified_count)
    print "New format:"
    mypp.pprint(db[COLLECTION_ELEMS].find_one(update_docs[0]["filter"]))

correct_chungking()
```

```python
# OSM-specific constants
HONGKONG_OSM = "hong-kong_china.osm"
KEY_ELEMENTS = ["bounds"]
CORE_ELEMENTS = ["node", "way", "relation"]
# regular expressions
REGX_LOWER = re.compile(r'^([a-z]|_)*$')
REGX_LOWER_COLON = re.compile(r'^([a-z]|_)*:([a-z]|_)*$')
REGX_LOWER_PROBLEMCHARS = re.compile(r'[=\+/&<>;\'"\?%#$@\,\. \t\r\n]')
```

```python
"""
Following the example to check feature keys
"""
def key_check(key, checks):
    lower_results = REGX_LOWER.search(key)
    lower_colon_results = REGX_LOWER_COLON.search(key)
    problemchars_results = REGX_LOWER_PROBLEMCHARS.search(key)

    if lower_results:
        checks["lower"] += 1
    elif lower_colon_results:
        checks["lower_colon"] += 1
    elif problemchars_results:
        checks["problemchars"] += 1
    else:
        checks["other"] += 1
        
    return checks
```

```python
"""
Extract the keys of attributes and sub-elements from OSM to see how contributors are logging in the area. 
And later I can avoid features that have insufficient information.
"""
def get_elements(source):
    # get an iterator
    context = et.iterparse(source, events=("start",))
    # get the root element
    event, root = context.next()
    
    elements = {}
    checks = {"lower": 0, "lower_colon": 0, "problemchars": 0, "other": 0}
    
    for event, elem in context:
        if event == "start":
            tag = elem.tag
            if tag not in elements:
                elements[tag] = {"count": 0, "attributes": []}
            elements[tag]["count"] += 1
            
            other_attrs = [x for x in elem.keys() if x not in elements[tag]["attributes"]]
            if other_attrs: # not empty
                if elements[tag]["attributes"]: # not empty
                    elements[tag]["attributes"].extend(other_attrs)
                else:
                    elements[tag]["attributes"] = other_attrs
            
            if tag in CORE_ELEMENTS:
                if "features" not in elements[tag]:
                    elements[tag]["features"] = {}
                for sub in elem.iter("tag"):
                    feature = sub.get("k")
                    key_check(feature, checks)
                    if feature not in elements[tag]["features"]:
                        elements[tag]["features"][feature] = 0
                    elements[tag]["features"][feature] += 1
                    
        # reset and clear the elements from memory
        root.clear()

    return elements, checks
```

```python
elements, checks = get_elements(HONGKONG_OSM)
pp.pprint(checks)
pp.pprint(elements)
```
